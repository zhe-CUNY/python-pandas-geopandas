```python
# filter for complaints in your zipcode using the api
ZIPCODE = "10012"
where_clause = f"created_date>='2026-06-01T00:00:00' AND incident_zip='{ZIPCODE}'"

url = (
    f"https://data.cityofnewyork.us/resource/erm2-nwe9.csv"
    f"?$limit=10000"
    f"&$where={quote(where_clause)}"
)

df = pd.read_csv(url)
```
